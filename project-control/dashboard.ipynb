# 赛博聊机 · 开发进度看板

读取 `dev_plan.json`，展示所有 Phase 状态、当前进展和下一步行动。

```python
import json
from datetime import datetime

with open('dev_plan.json', 'r', encoding='utf-8') as f:
    plan = json.load(f)

phases = plan['phases']
print(f"项目: {plan['project']}")
print(f"计划版本: v{plan['version']}  更新日期: {plan['updated']}")
print(f"共 {len(phases)} 个阶段")
```

## 1. 全部 Phase 总表

```python
STATUS_EMOJI = {'done': '✅', 'doing': '🔵', 'blocked': '🔴', 'todo': '⬜'}

print(f"{'Phase':<8} {'状态':<10} {'标题'}")
print('-' * 70)
for p in phases:
    icon = STATUS_EMOJI.get(p['status'], '?')
    print(f"Phase {p['phase']:<3} {icon} {p['status']:<8} {p['title']}")

done_count = sum(1 for p in phases if p['status'] == 'done')
print(f"\n进度: {done_count}/{len(phases)} ({done_count/len(phases)*100:.0f}%)")
```

## 2. 当前进行中 / 阻塞的 Phase

```python
active = [p for p in phases if p['status'] in ('doing', 'blocked')]

if not active:
    todo = [p for p in phases if p['status'] == 'todo']
    if todo:
        print(f"当前无进行中的 Phase。建议开始 Phase {todo[0]['phase']}: {todo[0]['title']}")
    else:
        print("🎉 所有 Phase 已完成!")
else:
    for p in active:
        icon = STATUS_EMOJI[p['status']]
        print(f"{icon} Phase {p['phase']}: {p['title']}")
        print(f"   状态: {p['status']}")
        print(f"   任务数: {len(p['tasks'])}")
        print()
```

## 3. 推荐模型

```python
target = active if active else [p for p in phases if p['status'] == 'todo'][:1]

for p in target:
    print(f"Phase {p['phase']}: {p['title']}")
    print(f"  设计阶段 (架构/接口边界): {p['model_for_design']}")
    print(f"  编码阶段 (实现/CRUD):      {p['model_for_build']}")
    print(f"  测试阶段 (跑测试/修bug):   {p['model_for_test']}")
    print()
```

## 4. 验收标准 & 测试项

```python
for p in target:
    print(f"=== Phase {p['phase']}: {p['title']} ===")
    print()
    print("验收标准:")
    for i, a in enumerate(p['acceptance'], 1):
        print(f"  {i}. {a}")
    print()
    print("要跑的测试:")
    for i, t in enumerate(p['tests_to_run'], 1):
        print(f"  {i}. {t}")
    print()
```

## 5. 需要你手动做的事

```python
for p in target:
    print(f"=== Phase {p['phase']}: {p['title']} ===")
    print()
    for i, item in enumerate(p['need_me'], 1):
        print(f"  ☐ {i}. {item}")
    print()
```

## 6. Next Prompt（复制给 AI 即可执行）

```python
for p in target:
    print(f"Phase {p['phase']} next_prompt:")
    print(f"\n  {p['next_prompt']}")
    print()
```
